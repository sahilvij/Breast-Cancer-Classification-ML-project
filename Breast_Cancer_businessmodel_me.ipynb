Breast Cancer Classification

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Importing Data

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data
```

```python
print(data['DESCR'])
```

```python
df = pd.DataFrame(np.c_[data['data'], data['target']], columns = np.append(data['feature_names'], ['targets']))
```

```python
df.head(9)
```

Visualizing

```python
sns.pairplot(df, hue = 'targets', vars = ['mean radius','mean texture', 'mean perimeter', 'mean compactness','mean smoothness'])
```

```python
sns.countplot(df['targets'])
```

```python
sns.scatterplot(x = 'mean perimeter',y = 'mean compactness', hue = 'targets', data = df)
```

```python
plt.figure(figsize = (20,14))
sns.heatmap(df.corr(), annot = True)
```

Model Training

```python
inputs = df.drop(['targets'], axis =1)
```

```python
target = df['targets']
```

```python
pd.options.display.max_rows = 999
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs_train, inputs_test, target_train, target_test = train_test_split(inputs, target, test_size=0.15, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = SVC()
```

```python
model.fit(inputs_train, target_train)
```

Evaluation

```python
predictions = model.predict(inputs_test)
```

```python
predictions
```

```python
cm = confusion_matrix(target_test, predictions)
```

```python
sns.heatmap(cm, annot=True)
```

Improving the Model

```python
print(classification_report(target_test, predictions))
```

```python
from sklearn import preprocessing
```

```python
scaled_inputs_train, scaled_inputs_test = preprocessing.scale(inputs_train), preprocessing.scale(inputs_test)
```

```python
model.fit(scaled_inputs_train, target_train)
```

```python
pred3 = model.predict(scaled_inputs_test)
```

```python
cm3 = confusion_matrix(target_test, pred3)
```

```python
sns.heatmap(cm3, annot = True)
```

```python
print(classification_report(target_test, pred3))
```

Modifying SVM parameters

```python
param_grid = {'C' : [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(SVC(), param_grid, refit = True, verbose =2)
grid_search.fit(scaled_inputs_train, target_train)
```

```python
grid_search.best_params_
```

```python
pred4 = grid_search.predict(scaled_inputs_test)
```

```python
cm4 = confusion_matrix(target_test, pred4)  
```

```python
sns.heatmap(cm4, annot=True)
```

```python
print(classification_report(target_test, pred4))
```
